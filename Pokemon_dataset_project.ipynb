## Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

## Problem Statement:

### You have been given a dataset that describe the functionality and features of differnet pokemons. Now based on the given information, you have to plot different graphs according to thier features.

```python
pokemon=pd.read_csv("Pokemon.csv")     #definig the datset
print(pokemon.shape)        # to check no. of rows and columns
pokemon.head()              # to print first five rows of a datset
```

## Bar Charts

```python
sb.countplot(data=pokemon, x="generation_id");      # to plot a bar graph on the basis of different generation_id on x-aixs
```

```python
base_color=sb.color_palette()[0]           # to print all the columns in a same colour.
sb.countplot(data=pokemon, x="generation_id", color=base_color)
```

```python
base_color=sb.color_palette()[1]                  # to plot the graph from maximum to minimum genertaion_id order
gen_order=pokemon["generation_id"].value_counts().index
sb.countplot(data=pokemon, x="generation_id", color=base_color, order=gen_order)
```

## Additional Variations:

```python
base_color=sb.color_palette()[2]          #to print the bar plot according to type1 column on x-axis
sb.countplot(data=pokemon, x="type_1",color=base_color)
```

```python
base_color=sb.color_palette()[2]                # to rotate the x-axis labels by 90 degrees
sb.countplot(data=pokemon, x="type_1", color=base_color)
plt.xticks(rotation=90);
```

```python
base_color=sb.color_palette()[2]            # to print the type1 column labels on y-axis
sb.countplot(data=pokemon, y="type_1",color=base_color)
```

## Count Missing Data:

```python
pokemon.isna().sum()            # to count the no. of empty values in each column
```

```python
na_counts=pokemon.isna().sum()            # to plot the bar
base_color=sb.color_palette()[5]
sb.barplot(na_counts.index.values, na_counts, color=base_color)
plt.xticks(rotation=90);
```

## Pie Charts

```python
sorted_counts= pokemon["generation_id"].value_counts()
plt.pie(sorted_counts, labels=sorted_counts.index, startangle=90, counterclock=False);
plt.axis("square");
```

## Additional Variation:

```python
sorted_counts=pokemon["generation_id"].value_counts()    #to cut a small circle of radius 0.4 from pie chart
plt.pie(sorted_counts, labels=sorted_counts.index, startangle=90, counterclock=False, wedgeprops={"width":0.4});
plt.axis("square");
```

## Histograms

```python
plt.hist(data=pokemon, x="speed")       # to create a histogram with speed labels on x-axis
```

```python
plt.hist(data=pokemon,x="speed",bins=20)     # to divide the total range of histogram into 20 parts
```

```python
bins=np.arange(0,pokemon["speed"].max()+1,5)              # to create bins in a range from 0 to maximum of speed
plt.hist(data=pokemon, x="speed", bins=bins)
```

```python
plt.figure(figsize=[10,5])           # to make larger figure size for subplots
#histogram on left, too-large bin size
plt.subplot(1,2,1)   #1 row, 2 cols, subplot 1
bin_edges=np.arange(0,pokemon["speed"].max()+1, +5)
plt.hist(data=pokemon, x="speed", bins=bin_edges)

#histogram on right, too-small bin size
plt.subplot(1,2,2)   #1 row, 2 cols, subplot 2
bin_edges=np.arange(0,pokemon["speed"].max()++1,1)
plt.hist(data=pokemon, x="speed", bins=bin_edges);
```

## Alternative Approach:

```python
sb.distplot(pokemon["speed"]);            # to show density or pokemons distribution with speed
```

```python
sb.distplot(pokemon["speed"],kde=False);        # to remove distribution of pokemons with speed
```

```python
bin_edges=np.arange(0,pokemon["speed"].max()+1,5)       # to increase the intensity of the histogram
sb.distplot(pokemon["speed"],bins=bin_edges,kde=False, hist_kws={"alpha":1});
```
